```python
import pandas as pd
```

```python
df=pd.read_csv("Random_Forest_Dataset.csv")
df.head()
```

```python
df.rename(columns={'Cost of Cultivation (`/Hectare) A2+FL':'column2','Cost of Cultivation (`/Hectare) C2':'Cultivation','Cost of Production (`/Quintal) C2':'production','Yield (Quintal/ Hectare) ':'yield'},inplace=True)
df.head()
```

```python
print(df.columns.tolist())
```

```python
df.drop(['State','column2','yield'],axis='columns',inplace=True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encode=df.copy()
encode.head()
```

```python
le_crop=LabelEncoder()
```

```python
encode['Crop_encode']=le_crop.fit_transform(df['Crop'])
encode.head()
```

```python
X=encode[['Crop_encode']]
y=encode['production']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
len(X_train)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(X_train, y_train)
```

```python
print("R² score:", model.score(X_test, y_test))
```

```python
y_predicted = model.predict(X_test)
```
